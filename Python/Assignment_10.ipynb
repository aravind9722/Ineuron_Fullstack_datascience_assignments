**1.How do you distinguish between shutil.copy() and shutil.copytree()?**

`shutil.copy(`) will copy a single file 

`shutil. copytree()` will copy an entire folder and every folder and file contained in it.

**3.What is the difference between the delete functions in the send2trash and shutil modules?**

`Shutil module` provides a funciton called as `shutil.rmtree()` which deletes a directory and all its contents. if the file is deleted it will be lost permanently, if a file is deleted accidentally using these methods there is no way we can recover the deleted file

`send2trash` module provides a function called `send2trash.send2trash()` to delete a file/directory. these methods moves the files/directories to trash folder instead of permanently deleting them. hence if a file/folder is deleted accidentally it can be still recovered from trash folder.

**5.Create a programme that searches a folder tree for files with a certain file extension (such as .pdf
or .jpg). Copy these files from whatever location they are in to a new folder.**

```python
import os, shutil

def selectionCopy(folder, extensions, destFolder):
	folder = os.path.abspath(folder)
	destFolder = os.path.abspath(destFolder)
	print('Looking in', folder, 'for files with extensions of', ', '.join(extensions))
	for foldername, subfolders, filenames in os.walk(folder):
		for filename in filenames:
			name, extension = os.path.splitext(filename)
			if extension in extensions:
				fileAbsPath = foldername + os.path.sep + filename
				print('Coping', fileAbsPath, 'to', destFolder)
				shutil.copy(fileAbsPath, destFolder)

extensions = ['.ipynb', '.jpg', '.pdf']
folder = 'randomFolder'
destFolder = 'selectiveFolder'
selectionCopy(folder, extensions, destFolder)
```
